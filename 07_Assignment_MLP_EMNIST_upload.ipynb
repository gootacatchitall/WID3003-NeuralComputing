That there are labels that our model is pretty confident in predicting about, but there are also labels that our model is still predicting (correctly) but not really sure of itself. **In this task, you are required to code and produce the results of unsure predictions for all labels**. Sample output is shown below.

Let's consider some label 'z'. Let's say that if our model is predicting 'z' with a probability of more than 90%, then surely we can say that our model is quite confident in predicting this particular label. But what if it is still predicting 'z' but with much lower probability? 

Let us set a threshold, if label 'z' is predicted with $p(z)\geq 0.6$ then the model is confident. Else it is a bit 'unsure'.
